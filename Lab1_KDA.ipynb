```python
import numpy as np
a=[428, 300, 250, 578, 492, 204, 245, 420, 760, 363, 600, 483, 711, 265, 572, 350, 652, 460, 340, 670, 565, 750, 452]
print("Последовательность а:\n", a, "\n")
print("Среднее значение последовательности а: %.2f"%np.mean(a))
print("Медиана последовательности а: %d"%np.median(a))
print("Стандартное отклонение последовательности а: %.2f"%np.std(a), "\n")

x = [428, 300, 250, 578, 492, 204, 245, 420, 760, 363, 600, 483, 711, 265, 572, 350, 652, 460, 340, 670, 565, 750, 452]
y = [21, 33, 43, 71, 85, 17, 38, 43, 53, 76, 33, 16, 24, 59, 65, 73, 45, 63, 53, 43, 58, 28, 37]
print("Последовательность x:\n", x)
print("Последовательность y:\n", y, "\n")
cov = (sum((x[k] - np.mean(x)) * (y[k] - np.mean(y)) for k in range(len(y)))/ ((len(y)) - 1))
print("Ковариация последовательностей х и у: %.2f"%cov)
print("Коэффициент Пирсона последовательностей х и у: %.4f"%np.corrcoef(x, y)[0, 1])
```
